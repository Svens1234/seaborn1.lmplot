```python
import seaborn as sns
```

```python
sns.set_theme(style="ticks")
```

```python
df = sns.load_dataset("anscombe")
```

```python
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, palette="muted", height=5,
           scatter_kws={"s": 9, "alpha": 0.8})
```
